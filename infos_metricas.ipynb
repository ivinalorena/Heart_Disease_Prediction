# Informaes sobre as bibliotecas utilizadas
---
- **Mtricas**:
    - ROC AUC: mtrica de avaliao em ML usada para avaliar o desempenho de modelos de classificao. Seu nome significa *rea Sob a Curva Caracterstica de Operao do Receptor*.
        - Como interpretar ROC AUC?

        Indica o quo bom seu modelo  em ranquear exemplos positivos acima dos exemplos negativos.
        ### Exemplo de como entender o ROC AUC:
        ---
        Imagine que voc tenha duas caixas, uma com um carto para cada um de seus exemplos positivos e outra com um carto para cada um de seus exemplos negativos.

        Em cada carto est escrito a pontuao que o modelo atribuiu deste exemplo ser positivo.

        Este nmero pode ser a probabilidade, mas isso no  obrigatrio, j que a ROC AUC  uma medida mais preocupada com o ranqueamento de exemplos do que com a preciso matemtica.

        

        Voc decide sortear um carto positivo e um negativo por 5 vezes, sempre recolocando os cartes sorteados de volta nas caixas.

        Aps o sorteio, voc faz uma tabela, indicando as probabilidades dos cartes sorteados cada vez que voc os retirou das caixas.

        Lembrando que as probabilidades no precisam somar um porque estamos falando de dois exemplos separados e no da probabilidade do mesmo exemplo ser da classe positiva ou negativa.

        

        Agora voc conta quantas vezes o exemplo positivo tinha uma pontuao maior do que o negativo e divide pelo nmero total de sorteios.
        Neste caso 4/5 = 0,8. Este  o valor do ROC AUC.

        **Em resumo**, quanto maior o valor do ROC AUC, melhor o modelo est ranqueando exemplos positivos acima dos negativos.

        Ele costuma variar de 0,5 (significa que o modelo no tem qualquer poder preditivo) a 1,0 (significa que o modelo prev as classes perfeitamente).

        Na prtica voc encontra nmeros menores que 0,5, indicando algum problema grave nos dados ou no algoritmo que voc treinou.

        ### Diferena entre ROC AUC e Acurcia
        ---
        **Acurcia**  uma mtrica muito mais simples: a porcentagem de classificaes corretas pelo modelo. Para calcul-la, voc precisa definir um ponto de corte para a probabilidade de um exemplo pertencer a uma das classes.

        Geralmente esse ponto de corte  0,5, ou seja, todos os exemplos com previses acima de 0,5 so considerados da classe positiva, enquanto os que estiverem abaixo, so atribudos  classe negativa.

        A **ROC AUC**, por outro lado, no requer um ponto de corte.

        Em vez disso, ela mede a capacidade do modelo de distinguir entre as duas classes em todos os pontos de corte possveis.

        Isso significa que ela  uma mtrica mais robusta da performance do modelo do que a acurcia, pois ela avalia o modelo sem depender de escolher o ponto de corte ideal.

        Pela mesma razo, a ROC AUC  uma mtrica melhor para casos em que as classes so bastante desbalanceadas (por exemplo, 10% de positivos e 90% de negativos).
