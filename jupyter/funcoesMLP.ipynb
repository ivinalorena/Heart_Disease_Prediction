### Explicando as funções na página do `PAGEMLP.py`

- _normalize: padroniza nomes de colunas (minúsculo + _) para aceitar CSV com variações de escrita e mapear para o formato esperado.
- _force_cpu_recursively: percorre o objeto do modelo e força device="cpu" (inclusive atributos internos), evitando chamadas CUDA.
- load_realmlp_model: carrega o .pkl com joblib, aplica o forçamento para CPU e usa cache de recurso do Streamlit.
- load_reference_data: lê o dataset de referência e transforma Heart Disease em numérico (LabelEncoder) para gerar as features estatísticas.
- add_engineered_features: recria as colunas orig_* (mean/median/std/skew/count) exatamente como no treino.
- prepare_model_input: valida colunas do upload, ordena as features, adiciona engenharia e converte para category, entregando o shape que o modelo espera.
- save_uploaded_dataframe: salva o CSV enviado em uploads com nome seguro e timestamp.
No bloco de submit, o try/except AssertionError faz fallback quando aparecer “Torch not compiled with CUDA enabled”.
